```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([6,7,8,9])
m=0
b=0
learningrate=0.01
epochs=1000
n=len(x)
for i in range(epochs):
    y_pred=m*x+b
    dm=(-2*n)/np.sum(x*(y-y_pred))
    db=(-2*n)/np.sum(y-y_pred)
    m=m-(learningrate*dm)
    n=n-(learningrate*db)
print("finalslope",m)
print("x-intercept",n)
plt.scatter(x,y,color='red')
plt.plot(x,m*x+b,y)
plt.show()
```
